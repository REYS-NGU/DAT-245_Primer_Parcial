# FOOD NUTRITION

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, KBinsDiscretizer, MinMaxScaler
```

```python
data_food = pd.read_csv("ABBREV.csv")
```

## Preprocesamiento OneHotEncoder, LabelEncoder, Discretización y Normalización

```python
onehotencoder = OneHotEncoder(sparse_output=False)
onehot_encoded = onehotencoder.fit_transform(data_food[['GmWt_Desc2']])
data_food_onehot = pd.DataFrame(onehot_encoded, columns=onehotencoder.get_feature_names_out(['GmWt_Desc2']))

data_food = pd.concat([data_food, data_food_onehot], axis=1)

data_food.head(21)
```

```python
labelencoder = LabelEncoder()
data_food['GmWt_Desc1_encoded'] = labelencoder.fit_transform(data_food['GmWt_Desc1'])
data_food.head(21)
```

```python
discretizer = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
data_food['Carbohydrt_discretized'] = discretizer.fit_transform(data_food[['Carbohydrt_(g)']])
data_food.head(21)
```

```python
scaler = MinMaxScaler()
data_food['Protein_normalized'] = scaler.fit_transform(data_food[['Protein_(g)']])
data_food.head(21)
```

```python
data_food.to_csv('ABBREV_Preprocesado.csv', index=False)

data_food.head(21)
```
