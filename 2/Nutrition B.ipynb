## B. Columnas relevantes

### Calorias:
- Ayuda a determinar cuánta energía contiene un alimento.
- Se puede utilizar para ver si un alimento es denso en calorías o no.
- Es relevante para personas que siguen dietas específicas (por ejemplo, perder peso o ganar masa muscular).
- Relacionada con la cantidad de macronutrientes (proteínas, grasas y carbohidratos).

### Grasas Totales (g):
- Es importante para saber cuánta grasa se consume con un alimento.
- Se puede dividir en grasa saturada, insaturada, etc.
- Es clave en el control de enfermedades cardiovasculares y para controlar niveles de colesterol.
- Relación con la densidad calórica de un alimento.

### Proteina (g):
- Es esencial para la reparación de tejidos, crecimiento muscular y muchas otras funciones biológicas.
- Útil para personas que buscan aumentar masa muscular o en dietas altas en proteínas.
- Permite ver si un alimento es rico en proteínas o no.
- Relación con otras macronutrientes para balance de la dieta.

### Carbohidratos (g):
- Son una fuente clave de energía, especialmente para actividades físicas.
- Importante para dietas enfocadas en rendimiento deportivo o control de peso.
- Se descomponen en azúcares, fibra, etc., lo que influye en el índice glucémico del alimento.
- Los alimentos ricos en carbohidratos a menudo están relacionados con mayores cantidades de calorías.


```python
selected_columns = data_food[
    ["Energ_Kcal", "Protein_(g)", "Lipid_Tot_(g)", "Carbohydrt_(g)"]
]

sns.set_theme(style="whitegrid")

pairplot = sns.pairplot(
    selected_columns, diag_kind="kde", markers="o", plot_kws={"alpha": 0.7, "s": 50}
)

plt.suptitle(
    "Relación entre Calorías, Grasa Total, Proteína y Carbohidratos",
    y=1.02,
    fontsize=16,
    fontweight="bold"
)

plt.show()
```

# Análisis de Nutrientes en el Dataset de Alimentos

## 1. Calorías (Energy_Kcal):
- **Rango de valores**: Las calorías varían significativamente en el gráfico, desde valores bajos (alrededor de 0-50 kcal) hasta valores altos (aproximadamente 800 kcal), lo que indica una amplia gama de alimentos, desde los bajos en calorías hasta los altamente calóricos.
- **Distribución**: La mayor parte de los alimentos están concentrados en un rango bajo a moderado de calorías, lo que sugiere que la mayoría de los productos en el dataset tienen entre 0 y 400 kcal.
- **Relación con otros nutrientes**: Existe una relación clara entre las calorías y la cantidad de grasa y carbohidratos en el alimento, ya que estos macronutrientes son las principales fuentes de energía.
- **Outliers**: Algunos puntos en el gráfico están en los extremos, lo que podría corresponder a alimentos muy densos calóricamente, como snacks procesados, comidas rápidas o productos muy grasos.

## 2. Proteínas (Protein_g):
- **Rango de valores**: Las proteínas varían entre 0 g y alrededor de 100 g, lo que sugiere una variedad de alimentos ricos en proteínas como carnes y legumbres, así como alimentos bajos en proteínas, como frutas y algunos productos procesados.
- **Distribución**: La mayoría de los alimentos tienen entre 0 y 50 g de proteínas, lo que es común en alimentos de consumo diario.
- **Correlación débil con calorías**: La relación entre calorías y proteínas no es muy fuerte. Algunos alimentos con muchas proteínas no tienen necesariamente muchas calorías, lo que es común en fuentes magras de proteínas como carnes magras y legumbres.
- **Outliers**: Existen algunos alimentos con niveles muy altos de proteínas que pueden corresponder a productos altamente concentrados en proteínas como suplementos, carnes rojas o productos derivados de proteínas animales.

## 3. Grasas Totales (Lipid_Tot_g):
- **Rango de valores**: Las grasas totales varían desde 0 g hasta alrededor de 100 g, con alimentos muy ricos en grasa como los aceites, mantequillas y ciertos alimentos procesados.
- **Distribución**: La mayoría de los alimentos tienen entre 0 y 50 g de grasa, con pocos alimentos superando ese valor. Esto refleja que la mayoría de los alimentos tienen un contenido moderado de grasa.
- **Relación fuerte con calorías**: La correlación entre grasa y calorías es bastante clara, ya que los alimentos con más grasa aportan significativamente más calorías, lo cual es coherente con la densidad calórica de la grasa (9 kcal por gramo).
- **Outliers**: Los alimentos con más de 50 g de grasa son menos comunes pero representan productos que son principalmente grasa, como aceites, frutos secos y alimentos fritos.

## 4. Carbohidratos (Carbohydrt_g):
- **Rango de valores**: Los carbohidratos varían de 0 g a alrededor de 100 g, con una amplia gama de alimentos, desde aquellos casi sin carbohidratos (como las carnes) hasta alimentos ricos en carbohidratos (como panes, cereales, frutas y productos horneados).
- **Distribución**: La mayoría de los alimentos contienen entre 0 y 50 g de carbohidratos, lo que refleja una gama de alimentos comunes como verduras, productos lácteos y granos.
- **Relación clara con calorías**: Existe una correlación positiva entre carbohidratos y calorías, ya que los alimentos con muchos carbohidratos aportan más calorías (4 kcal por gramo de carbohidrato).
- **Outliers**: Algunos alimentos extremadamente altos en carbohidratos (más de 50 g) probablemente representan productos como dulces, productos horneados, y alimentos con mucho almidón.


El dataset de Kaggle titulado **"The Nutritional Content of Food: A Comprehensive Dataset"** contiene 54 columnas que representan varias características relacionadas con el contenido nutricional de diferentes alimentos.

### Columnas:

- **index**: Identificador único para cada fila.
- **NDB_No**: Número de identificación del alimento en la base de datos NDB (National Nutrient Database de USDA).
- **Shrt_Desc**: Descripción corta del alimento.
- **Water_(g)**: Cantidad de agua presente en gramos por cada 100 g del alimento.
- **Energ_Kcal**: Energía en calorías por cada 100 g del alimento.
- **Protein_(g)**: Cantidad de proteína en gramos por cada 100 g.
- **Lipid_Tot_(g)**: Total de lípidos (grasas) en gramos.
- **Ash_(g)**: Ceniza, que representa los minerales inorgánicos presentes.
- **Carbohydrt_(g)**: Carbohidratos en gramos por cada 100 g.
- **Fiber_TD_(g)**: Fibra dietética total en gramos.
- **Sugar_Tot_(g)**: Azúcar total en gramos.
- **Calcium_(mg)**: Calcio en miligramos.
- **Iron_(mg)**: Hierro en miligramos.
- **Magnesium_(mg)**: Magnesio en miligramos.
- **Phosphorus_(mg)**: Fósforo en miligramos.
- **Potassium_(mg)**: Potasio en miligramos.
- **Sodium_(mg)**: Sodio en miligramos.
- **Zinc_(mg)**: Zinc en miligramos.
- **Copper_(mg)**: Cobre en miligramos (la columna parece estar mal etiquetada como *Copper_mg*).
- **Manganese_(mg)**: Manganeso en miligramos.
- **Selenium_(µg)**: Selenio en microgramos.
