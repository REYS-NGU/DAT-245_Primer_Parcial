### C. MEDIDAS DE TENDENCIA CENTRAL

```python
selected_columns = data_food[["Energ_Kcal",'Protein_(g)', 'Lipid_Tot_(g)', 'Carbohydrt_(g)']]

media = selected_columns.mean()
mediana = selected_columns.median()
moda = selected_columns.mode().iloc[0]

print("Media:\n", media)
print("\nMediana:\n", mediana)
print("\nModa:\n", moda)
```

```python
selected_columns = data_food[["Energ_Kcal", 'Protein_(g)', 'Lipid_Tot_(g)', 'Carbohydrt_(g)']]

# Crear subgrÃ¡ficos
fig, axs = plt.subplots(1, 4, figsize=(12, 24))  # 1 fila, 4 columnas

# GrÃ¡fico de caja y bigotes para Energ_Kcal
sns.boxplot(data=selected_columns[["Energ_Kcal"]], ax=axs[0], color='lightblue')  # Usar el primer subplot
axs[0].set_title('CalorÃ­as (Energ_Kcal)')
axs[0].set_ylabel('Kcal')
axs[0].set_yticks(range(0, selected_columns["Energ_Kcal"].max() + 10, 20))

# GrÃ¡fico de caja y bigotes para ProteÃ­nas
sns.boxplot(data=selected_columns[['Protein_(g)']], ax=axs[1], color='salmon')  # Usar el segundo subplot
axs[1].set_title('ProteÃ­nas (g)')
axs[1].set_ylabel('Cantidad en gramos')
axs[1].set_yticks(range(0, 101, 5))  

# GrÃ¡fico de caja y bigotes para Grasas
sns.boxplot(data=selected_columns[['Lipid_Tot_(g)']], ax=axs[2], color='lightgreen')  # Usar el tercer subplot
axs[2].set_title('Grasas (g)')
axs[2].set_ylabel('Cantidad en gramos')
axs[2].set_yticks(range(0, 101, 10))  

# GrÃ¡fico de caja y bigotes para Carbohidratos
sns.boxplot(data=selected_columns[['Carbohydrt_(g)']], ax=axs[3], color='orange')  # Usar el cuarto subplot
axs[3].set_title('Carbohidratos (g)')
axs[3].set_ylabel('Cantidad en gramos')
axs[3].set_yticks(range(0, 101, 10))  

# Ajustar el espacio entre grÃ¡ficos
plt.tight_layout()

plt.show()
```

### Energ_Kcal (CalorÃ­as):
- El grÃ¡fico muestra una distribuciÃ³n de calorÃ­as donde la mayorÃ­a de los valores se concentran entre 100 y 340 kcal, con una mediana cercana a 191 kcal.

### Macronutrientes (ProteÃ­nas, Grasas y Carbohidratos):

#### Para las ProteÃ­nas:

- La distribuciÃ³n tambiÃ©n estÃ¡ sesgada hacia valores bajos, con la mayorÃ­a de los alimentos teniendo menos de 10 gramos de grasa total.

#### Para los Carbohidratos:
- La mayorÃ­a de los alimentos tienen entre 0 y 35 gramos de carbohidratos con una mediana de 9.34.
